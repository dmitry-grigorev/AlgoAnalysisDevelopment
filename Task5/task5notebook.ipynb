```python
from task5module.task5module import *
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
n = 100
m = 200

np.random.seed(42)
initial = np.random.permutation([1]*m+[0]*(n*(n-1)//2-m))
Madj = np.zeros(shape = (n, n), dtype = "int")
l = n-1
start = 0
end = l
for i in range(1, n):
    Madj[i-1, i:] = initial[start:end]
    l = n-i
    start = end
    end += l-1
Madj += Madj.T
```
