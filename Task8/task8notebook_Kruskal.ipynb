```python
from task8module.task8module import *
from Task6.task6module.task6module import *
import numpy as np
from typing import List
import copy
```

```python
weights = [10., 20., 30., 50., 100., 300.]
```

```python
np.random.seed(42)
n = 5
Madj = generate_from_nm_weighted(n, 7, weights)
graph1 = Graph(Madj.tolist())
```

```python
import networkx as nx
import matplotlib.pyplot as plt
g = nx.from_numpy_matrix(Madj)
np.random.seed(429)
plt.figure(figsize=(4, 4))
nx.draw(g, node_size = 100, pos = nx.circular_layout(g),  with_labels = True )
labels = nx.get_edge_attributes(g,'weight')
nx.draw_networkx_edge_labels(g,pos = nx.circular_layout(g),edge_labels=labels)
```

```python
graph1MST, cost1 = graph1.KruskalMST()
cost1
```

```python
graph1MSTgraph = nx.Graph()
graph1MSTgraph.add_nodes_from(g.nodes(data=True) )
graph1MSTgraph.add_weighted_edges_from([(x[0],x[1],x[2]) for x in sorted(graph1MST,
                            key=lambda item: item[0])])
np.random.seed(429)
plt.figure(figsize=(4, 4))
nx.draw(graph1MSTgraph, node_size = 100, pos = nx.circular_layout(graph1MSTgraph),  with_labels = True )
labels = nx.get_edge_attributes(graph1MSTgraph,'weight')
nx.draw_networkx_edge_labels(graph1MSTgraph,pos = nx.circular_layout(graph1MSTgraph),edge_labels=labels)
```
